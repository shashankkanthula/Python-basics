Operators
  Arithmetic

```python
num1=5
num2=10
print(num1+num2)
print(num1-num2)
print(num1*num2)
print(num1/num2)
print(num1%num2)
print(num1**num2)
print(num1//num2)
```

Assignment Operators

```python
num1=10
print(num1)
num1+=5
print(num1)
num1-=5
print(num1)
num1*=5
print(num1)
num1/=5
print(num1)
num1%=5
print(num1)
```

Comparison Operator


```python
num1=5
num2=8
print(num1==num2)
print(num1!=num2)
print(num1>num2)
print(num1<num2)
print(num1>=num2)
print(num1<=num2)
```

Logical

```python
num1=10
print(num1>20 and num1>5)
print(num1>20 or num1>5)
print(not(num1>8 and num1>5))
```

Identity

```python
a=["apple","banana"]
b=["apple","banana"]
c=a
print(a is c)
print(a is b)
print(a==b)
```

Bitwise

```python
num1=10
num2=4
print(num1&num2)
print(num1|num2)
print(num1^num2)
print(~num2)
print(num1<<2)
print(num1>>2)
```
