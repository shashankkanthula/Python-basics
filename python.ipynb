```python
num1=5
num2=10
print(num1+num2)
print(num1-num2)
print(num1*num2)
print(num1/num2)
print(num1%num2)
print(num1**num2)
print(num1//num2)
```

Assignment Operators

```python
num1=10
print(num1)
num1+=5
print(num1)
num1-=5
print(num1)
num1*=5
print(num1)
num1/=5
print(num1)
num1%=5
print(num1)
```

Comparison Operator


```python
num1=5
num2=8
print(num1==num2)
print(num1!=num2)
print(num1>num2)
print(num1<num2)
print(num1>=num2)
print(num1<=num2)
```

Logical

```python
num1=10
print(num1>20 and num1>5)
print(num1>20 or num1>5)
print(not(num1>8 and num1>5))
```

Identity

```python
a=["apple","banana"]
b=["apple","banana"]
c=a
print(a is c)
print(a is b)
print(a==b)
```

Bitwise

```python
num1=10
num2=4
print(num1&num2)
print(num1|num2)
print(num1^num2)
print(~num2)
print(num1<<2)
print(num1>>2)
```

slicing


```python
b="Arrangement of indexing examples"
print(b[2:5]) #we define (last indexing is not included)
print(b[:5])  #from start
print(b[2:])  #from end
print(b[-5:-2]) #negative indexing
```

modify strings

```python
a="    Bablu , b"
print(a.upper())
print(a.lower())
print(a.strip()) #removes whitespace at beg or end
print(a.replace("b","k")) #camelcase also sensitive
print(a.split(","))
```
